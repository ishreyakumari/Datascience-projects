Plot the Pie chart between Indian cities vs No. of jobs opening.
Print the Indian cities and %age of Job distribution in India upto 2 decimal places.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def function(location):
    loc=location.split(",")
    return pd.Series(loc)    

amazon=pd.read_csv("amazon_jobs_dataset.csv")
df=amazon.copy()
series=df["location"].apply(function)
df["Country"]=series[0]
df=df[df["Country"]=="IN"]
df["City"]=series[2]
city=df["City"].value_counts()
city_name=city.index
count=city.values
for i in range(city.size):
    print(city_name[i],count[i])
plt.pie(city,labels=city_name,autopct="%.2f")
plt.show()
```
