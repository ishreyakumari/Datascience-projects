Plot the line graph between no. of Job postings with respect to year.
Print the year and the number of job posting as integer value.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def function(date):
    return (date.split(",")[1]).strip()
amazon=pd.read_csv("amazon_jobs_dataset.csv")
df=amazon.copy()
df["Year"]=df["Posting_date"].apply(function)
amazon_period=df["Year"].value_counts()
amazon_year=amazon_period.index
amazon_year=amazon_year[-1::-1]
amazon_count=amazon_period.values
amazon_count=amazon_count[-1::-1]
for i in range(amazon_year.size):
    print(amazon_year[i],amazon_count[i])
plt.plot(amazon_year,amazon_count)
plt.xlabel("Year")
plt.ylabel("Number of jobs")
plt.yticks(np.arange(0,1750,100))
plt.grid()
plt.show()
```
