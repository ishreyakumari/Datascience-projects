Find total number of casualties in each country.

```python
import csv
with open("year2017.csv") as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    d_casualties={}
    for i in file_data:
        if(i["Country"] in d_casualties.keys()):
            d_casualties[i["Country"]]+=1
        else:
            d_casualties[i["Country"]]=1
    for k,v in d_casualties.items():
        print(k,v)
```
