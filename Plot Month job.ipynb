Plot the Bar graph between Month vs Job Openings.
Print the month name and the number of job posting as integer value.

```python
import pandas as pd
import matplotlib.pyplot as plt
def function(date):
    return date.split(" ")[0].strip()
amazon=pd.read_csv("amazon_jobs_dataset.csv")
df=amazon.copy()
df["Month"]=df["Posting_date"].apply(function)
job=df["Month"].value_counts()
month=job.index
count=job.values
for i in range(month.size):
    print(month[i],count[i])
plt.bar(month,count,)
plt.xticks(rotation=40)
plt.show()
```
