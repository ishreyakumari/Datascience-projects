Find out the percentage of the total earning for each genre in IMDB Movie Dataset?Plot the pie chart.
Print the genre and percentage of movies total earning present in that genre with 2 decimal place after rounding off.

```python
import pandas as pd
import sqlite3 as sql
import matplotlib.pyplot as plt
import numpy as np
db=sql.connect("IMDB.sqlite")
imdb=pd.read_sql_query("select genre,Domestic,Worldwide from earning e join imdb i on i.movie_id=e.movie_id join genre g on g.movie_id=i.movie_id",db)
df=imdb.copy()
df=df[df["genre"]!=""]
df["Domestic"].fillna(0.0)
df["Worldwide"].fillna(0.0)
df["Earning"]=df["Domestic"]+df["Worldwide"]
movie= df.groupby("genre")["Earning"].sum()
genre=movie.index
earning=movie.values
per= np.true_divide(earning,earning.sum())*100
for i in range(per.size):
    print(genre[i],"{:.2f}".format(per[i]))
plt.pie(earning,labels=genre,autopct="{:.2f}")
plt.show()
```
